```python
###Checking whether there is a correlation or not
sns.barplot(x=df["study_hours"],y=df["exam_score"], palette=["red", "blue"], hue=df["placement_status"])
```

```python
### Checking whether Study hour affect on score
sns.barplot(x=df["placement_status"],y=df["study_hours"], color="orange")
```

```python
sns.lineplot(x=df["study_hours"],y=df["placement_status"], color="green")
```

```python
sns.countplot(data=df,x="placement_status")
```

```python
###Percentage of placed using piechart visual 
counts = df["placement_status"].value_counts()

plt.pie(
    counts,
    labels=counts.index,
    autopct="%1.1f%%",
    colors=["lightgreen", "blue"],
    startangle=90
)

plt.title("Placement Status Distribution")
plt.show()
```
